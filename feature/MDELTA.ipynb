```python
import os
l = []
for i in myargs:
    l.append(myargs[i].value)
mdelta = '/home/ee_while/mdelta/mdelta/mDELTA.py'
if l[22].strip() == '' or l[22].strip() == '/': # 如果为空或者为 / 的集合，默认为当前目录
    output = os.getcwd() + '/'
elif l[22][0] == '/': # 如果是绝对路径
    if l[22][-1] != '/': 
        output = l[22]+ '/'
    else: output = l[22]
else: # 相对路径
    if l[22][-1] != '/': 
        output = os.getcwd() + '/' + l[22] + '/'
    else: output = os.getcwd() + '/' + l[22]
    
    
    
    
for pp in forlist(l[7][0], l[7][1], l[8]):
    for ma in forlist(l[10][0], l[10][1], l[11]):
        for mi in forlist(l[12][0], l[12][1], l[13]):
            for p in forlist(l[14][0], l[14][1], l[15]):
                for mg in forlist(l[18][0], l[18][1], l[19]):
                    for x in forlist(l[20][0], l[20][1], l[21]):
                        # 消除红色提示
                        os.system('python3 ' + mdelta + ' {0[0]} {0[1]}'.format(l)
                                  + (' -nt {0[2]}'.format(l) if len(l[2])!=0 else '')
                                  + (' -nt2 {0[3]}'.format(l) if len(l[3])!=0 else '')
                                  + (' -sd {0[6]}'.format(l) if len(l[6])!=0 else '')
                                  + (' -sd1 {0[4]}'.format(l) if len(l[4])!=0 else '')
                                  + (' -sd2 {0[5]}'.format(l) if len(l[5])!=0 else '')
                                  + ' -pp ' + str(pp)
                                  + ' -t {0[9]}'.format(l)
                                  + ' -ma ' + str(ma)
                                  + ' -mi ' + str(mi)
                                  + ' -p ' + str(p)
                                  + ' -n' + (' 1' if l[17] else ' 0')
                                  + ' -T' + (' 1' if l[16] else ' 0')
                                  + ' -mg ' + str(mg)
                                  + ' -x ' + str(x)
                                  + (' -o {0[22]}'.format(l) if len(l[22])!=0 else '')
                                  + ' -P {0[23]}'.format(l)
                                  + ' -c {0[24]}'.format(l)
                                 )
                        
                        result_file = output + '{}_{}_top{}_diff{}_pv{}_pp{}_miv{}_mav{}_mg{}.json'.format(
                            os.path.basename(l[0]).split('.')[0], 
                            os.path.basename(l[1]).split('.')[0],
                            l[9], str(x), str(p), str(pp), str(mi), str(ma), str(mg))
                        
                        name2type1 = l[2]
                        name2type2 = l[3]
                        scoredict = l[6]
                        if name2type1.strip() == '' : name2type1='non'
                        if name2type2.strip() == '' : name2type2='non'
                        if scoredict.strip() == '' : scoredict='non'
                        
                        %run /home/ee_while/JOB230605/draw.py --list $result_file $output $name2type1 $name2type2 $scoredict $ma
                        

print('mDELTA OK!!!')
```
